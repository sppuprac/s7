```python
df.drop(columns=['ADDRESSLINE1','ADDRESSLINE2','STATUS','POSTALCODE','STATE','TERRITORY','CITY','ORDERNUMBER','CUSTOMERNAME','PHONE','CONTACTLASTNAME','CONTACTFIRSTNAME','ORDERDATE'],axis=1, inplace=True)
```

```python
DEALSIZE = pd.get_dummies(df['DEALSIZE'])
```

```python
DEALSIZE
```

```python
PRODUCTLINE = pd.get_dummies(df['PRODUCTLINE'])
COUNTRY = pd.get_dummies(df['COUNTRY'])
```

```python
df = pd.concat([df,DEALSIZE,PRODUCTLINE,COUNTRY], axis=1)
```

```python
df
```

```python
df.drop(columns=['DEALSIZE','PRODUCTLINE','COUNTRY'],axis=1, inplace=True)
```

```python
df
```

```python
df['PRODUCTCODE'] = pd.Categorical(df['PRODUCTCODE']).codes
```

```python
df
```

```python
df.dtypes
```

```python
from sklearn.cluster import KMeans
```

```python
distortion = []
k = range(1,10)
for n in k:
    km = KMeans(n_clusters=n)
    km.fit(df)
    distortion.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(k,distortion,'bx-')
plt.xlabel('k')
plt.ylabel('distortion')
plt.title('k vs distortion')
plt.show()
```

```python
x_train = df.values
model = KMeans(n_clusters=4, random_state=2)
model.fit(x_train)
```

```python
pred = model.predict(x_train)
```
