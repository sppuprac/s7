```python
import pandas as pd
```

```python
df = pd.read_csv("Churn_Modelling.csv")
df
```

```python
df.drop(columns = ['RowNumber','CustomerId','Surname'], inplace=True)
```

```python
df
```

```python
geography = pd.get_dummies(df['Geography'])
gender = pd.get_dummies(df['Gender'])
```

```python
df = pd.concat([df,geography,gender],axis=1)
df
```

```python
df.drop(columns=['Geography','Gender'],inplace=True)
```

```python
df
```

```python
X = df.drop(['Exited'],axis=1)
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
classifier.add(Dense(activation='relu', input_dim=13, units=6, kernel_initializer='uniform')) #input layer
classifier.add(Dense(activation='relu', units=6, kernel_initializer='uniform')) #hidden layer
classifier.add(Dense(activation='sigmoid', units=1)) #output layer
#sigmoid for binary classification and softmax for more than 2 classification
classifier.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
classifier.summary()
```

```python
classifier.fit(x_train,y_train, epochs=50, batch_size=10)
```

```python
F_pred = classifier.predict(x_test)
```

```python
y_pred = (F_pred>0.5)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```

```python
precision = precision_score(y_test,y_pred)
precision
```

```python
recall = recall_score(y_test,y_pred)
recall
```

```python
f1= f1_score(y_test,y_pred)
f1
```
